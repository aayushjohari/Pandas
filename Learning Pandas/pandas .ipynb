```python
import pandas as pd
import numpy as np
```

```python
lst =[1,2,3,4,5]

pd.Series(lst)
```

## giving index from our own end

```python
pd.Series(data = lst , index = [1,2,3,4, 5])
```

## Series through dictionary values

```python
steps ={'day1': 4000 , "day2": 5000}
pd.Series(steps)
```

## Using repeat functions

```python
pd.Series(5).repeat(3)
```

## AGGREGATE FUNCTIONS

```python

sr = pd.Series([1,2,3,4,5,6])
sr.agg(["max", "min"])
```

## APPENDING FUNCTIONS

```python
sr1 = pd.Series([1,-2,3])
sr2 = pd.Series([1,2,3])
pd.concat([sr1,sr2])
```
