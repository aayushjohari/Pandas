## filling null values - Imputation

```python
data3 = data.copy()
```

```python
data3
```

```python
data3.info()
```

```python
data3.isnull().sum()
```

```python
data3['reviews_per_month']
```

```python
median_value = data3['reviews_per_month'].median()
```

```python
data3['reviews_per_month'] = round(data3['reviews_per_month'].fillna(median_value),2)
```

```python
data3
```

```python
data3.isnull().sum()
```

## categorical data can be handle in 2 ways:
1 - by using mode
2-  make new category

## by using mode

```python
data3['last_review']=data3['last_review'].fillna(data3['last_review'].mode()[0])
```

```python
data3
```

```python
data3.isnull().sum()
```

```python
data3.shape
```

## making new category

```python
data4 = data.copy()
```

```python
data4['last_review']=data4['last_review'].fillna("Not reviewed")
```

```python
data4
```

## advanced imputation

```python
data5 = data.copy()
```

```python
data5['reviews_per_month']=data5['reviews_per_month'].interpolate()
```
