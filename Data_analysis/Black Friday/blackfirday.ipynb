## ANALYSE GENDER

```python
sns.countplot(data['Gender'])
```

```python
data['Gender'].value_counts()
```

```python
plt.title('Gender analysis')
plt.pie(x = data['Gender'].value_counts() ,labels= data['Gender'].value_counts().index ,  autopct= '%.2f%%' ,  wedgeprops={'edgecolor' : "black"})
plt.legend()
plt.show()
```

```python
plt.bar(x = data['Gender'].value_counts().index , height=data['Gender'].value_counts().values)
plt.title("Gender analysis")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
amountspend =data.groupby('Gender')['Purchase'].sum()
```

```python
amountspend
```

```python
plt.bar(amountspend.index , amountspend.values)
plt.show()
```

```python
plt.pie(x = amountspend.values , autopct='%.2f%%' , labels =  amountspend.index)
plt.title("Amount spend by male and female")
plt.legend()
plt.show()
```

```python
averageamount = data.groupby('Gender')['Purchase'].mean()
```

```python
averageamount
```

```python
plt.pie(x = averageamount.values , autopct='%.2f%%' , labels =  averageamount.index)
plt.title("Amount spend by per male and female")
plt.legend()
plt.show()
```
