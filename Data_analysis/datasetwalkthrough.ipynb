## Q2. How many apps are there with rating 5?

```python
count = 0
for i in data['Rating']:
    if i == 5.0:
        count+=1

print(f"There are {count} apps having rating 5")
```

## Q3. How many apps are there with rating between 4 - 4.5?


```python
count = 0 
for i in data['Rating']:
    if i >= 4 and i <= 4.5:
        count+=1

print(f"There are {count} having rating between 4-4.5")
```

## Total number of apps in each category

```python
for name in data['Category'].unique():
    print(name)
```

```python
categories = {}

for name in data['Category'].unique():
    count =0 
    for i in data['Category']:
        if i == name :
            count+=1
    categories[name] = count

for key, value in categories.items():
    print(key,':', value)
```

```python
data['Category'].value_counts()
```

```python
categories = {}

for category in data['Category']:
    if category in categories:
        categories[category] += 1
    else:
        categories[category] = 1

for key, value in categories.items():
    print(key,':' , value)
```
