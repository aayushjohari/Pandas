## find the average rating 

```python
avg_rating=round(int(sum(data['Rating']))/len(data['Rating']) , 2)
```

```python
print("The average rating of the apps  is: ", avg_rating)
```

## Q2. How many apps are there with rating 5?

```python
count = 0
for i in data['Rating']:
    if i == 5.0:
        count+=1

print(f"There are {count} apps having rating 5")
```

## Q3. How many apps are there with rating between 4 - 4.5?


```python
count = 0 
for i in data['Rating']:
    if i >= 4 and i <= 4.5:
        count+=1

print(f"There are {count} having rating between 4-4.5")
```

## Q4. Average App Reviews


```python
data['Reviews']
```

```python
s = 0
for i in data['Reviews']:
    s+=int(i)

avg_reviews = int(s/len(data['Reviews']))
```

```python
print(f"The average of the app reveiws: ", avg_reviews)
```
